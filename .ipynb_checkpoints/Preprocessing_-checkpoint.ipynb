# Sound Anomaly Use Case (Preprocessing)
A jupyter notebook where it will do the preprocessing and extract all the features.

First is to import all the necessary libraries, then get the csv with the filepath of the sound files.

```python
import numpy as np
import pandas as pd
import librosa
import time
```

```python
start_time = time.time()
```

```python
df = pd.read_csv('datasets/filepath_valve_6dB.csv')
time
```

```python
df
```

Then we extract all the features.

```python
def get_y_sr(filename):
    y, sr = librosa.load(filename)
    return  y, sr
```

```python
df['y and SR'] = df['File Path'].apply(get_y_sr)
```

```python
df['y'] = df['y and SR'].apply(lambda x: x[0])
df['SR'] = df['y and SR'].apply(lambda x: x[1])
```

```python
df['y mean'] = df['y'].apply(lambda x : x.mean())
df['y max'] = df['y'].apply(lambda x: x.max())
df['y min'] = df['y'].apply(lambda x: x.min())
```

```python
def chroma_stft(y):
    chroma_stft = librosa.feature.chroma_stft(y=y[0], sr=y[1])
    return chroma_stft
```

```python
df['chroma_stft'] = df['y and SR'].apply(chroma_stft)
```

```python
df['chroma_stft min'] = df['chroma_stft'].apply(lambda x: x.min())
df['chroma_stft mean'] = df['chroma_stft'].apply(lambda x: x.mean())
df['chroma_stft max'] = df['chroma_stft'].apply(lambda x: x.max())
```

```python
def chroma_cqt(y):
    chroma_cqt = librosa.feature.chroma_cqt(y=y[0], sr=y[1])
    return chroma_cqt
```

```python
df['chroma_cqt'] = df['y and SR'].apply(chroma_cqt)
```

```python
df['chroma_cqt min'] = df['chroma_cqt'].apply(lambda x: x.min())
df['chroma_cqt mean'] = df['chroma_cqt'].apply(lambda x: x.mean())
df['chroma_cqt max'] = df['chroma_cqt'].apply(lambda x: x.max())
```

```python
df.head()
```

```python
df['chroma_cens'] = df['y'].apply(lambda x : librosa.feature.chroma_cens(y=x, sr=22050))
```

```python
df['chroma_cens min'] = df['chroma_cens'].apply(lambda x: x.min())
df['chroma_cens mean'] = df['chroma_cens'].apply(lambda x: x.mean())
df['chroma_cens max'] = df['chroma_cens'].apply(lambda x: x.max())
```

```python
df['melspectogram'] = df['y and SR'].apply(lambda x: librosa.feature.melspectrogram(y=x[0], sr=x[1]))
```

```python
df['melspectogram min'] = df['melspectogram'].apply(lambda x: x.min())
df['melspectogram mean'] = df['melspectogram'].apply(lambda x: x.mean())
df['melspectogram max'] = df['melspectogram'].apply(lambda x: x.max())
```

```python
df['mfcc'] = df['y and SR'].apply(lambda x: librosa.feature.mfcc(y=x[0], sr=x[1]))
```

```python
df['mfcc min'] = df['mfcc'].apply(lambda x: x.min())
df['mfcc mean'] = df['mfcc'].apply(lambda x: x.mean())
df['mfcc max'] = df['mfcc'].apply(lambda x: x.max())
```

```python
df['rms'] = df['y'].apply(lambda x: librosa.feature.rms(y=x))
```

```python
df['rms min'] = df['rms'].apply(lambda x: x.min())
df['rms mean'] = df['rms'].apply(lambda x: x.mean())
df['rms max'] = df['rms'].apply(lambda x: x.max())
```

```python
df['spectral_centroid'] = df['y and SR'].apply(lambda x: librosa.feature.spectral_centroid(y=x[0], sr=x[1]))
```

```python
df['spectral_centroid min'] = df['spectral_centroid'].apply(lambda x: x.min())
df['spectral_centroid mean'] = df['spectral_centroid'].apply(lambda x: x.mean())
df['spectral_centroid max'] = df['spectral_centroid'].apply(lambda x: x.max())
```

```python
df['spec_bw'] = df['y and SR'].apply(lambda x: librosa.feature.spectral_bandwidth(y=x[0], sr=x[1]))
```

```python
df['spec_bw min'] = df['spec_bw'].apply(lambda x: x.min())
df['spec_bw mean'] = df['spec_bw'].apply(lambda x: x.mean())
df['spec_bw max'] = df['spec_bw'].apply(lambda x: x.max())
```

```python
def spec_contrast(y):
    S = np.abs(librosa.stft(y[0]))
    contrast = librosa.feature.spectral_contrast(S=S, sr=y[1])
    return contrast
```

```python
df['spec_contrast'] = df['y and SR'].apply(spec_contrast)
```

```python
df['spec_contrast min'] = df['spec_contrast'].apply(lambda x: x.min())
df['spec_contrast mean'] = df['spec_contrast'].apply(lambda x: x.mean())
df['spec_contrast max'] = df['spec_contrast'].apply(lambda x: x.max())
```

```python
df['flatness'] = df['y'].apply(lambda x: librosa.feature.spectral_flatness(y=x))
```

```python
df['flatness min'] = df['flatness'].apply(lambda x: x.min())
df['flatness mean'] = df['flatness'].apply(lambda x: x.mean())
df['flatness max'] = df['flatness'].apply(lambda x: x.max())
```

Approximate maximum frequencies with roll_percent=0.85 (default)

```python
df['rolloff'] = df['y and SR'].apply(lambda x: librosa.feature.spectral_rolloff(y=x[0], sr=x[1]))
```

```python
df['rolloff min'] = df['rolloff'].apply(lambda x: x.min())
df['rolloff mean'] = df['rolloff'].apply(lambda x: x.mean())
df['rolloff max'] = df['rolloff'].apply(lambda x: x.max())
```

Fit a degree-0 polynomial (constant) to each frame [degree = 0]

```python
def poly_0(y):
    S = np.abs(librosa.stft(y[0]))
    p0 = librosa.feature.poly_features(S=S, order=0)
    return p0
```

```python
df['poly_0'] = df['y and SR'].apply(poly_0)
```

```python
df['poly_0 min'] = df['poly_0'].apply(lambda x: x.min())
df['poly_0 mean'] = df['poly_0'].apply(lambda x: x.mean())
df['poly_0 max'] = df['poly_0'].apply(lambda x: x.max())
```

Fit a linear polynomial to each frame [degree = 1]


```python
def poly_1(y):
    S = np.abs(librosa.stft(y[0]))
    p1 = librosa.feature.poly_features(S=S, order=1)
    return p1
```

```python
df['poly_1'] = df['y and SR'].apply(poly_1)
```

```python
df['poly_1 min'] = df['poly_1'].apply(lambda x: x.min())
df['poly_1 mean'] = df['poly_1'].apply(lambda x: x.mean())
df['poly_1 max'] = df['poly_1'].apply(lambda x: x.max())
```

Fit a quadratic to each frame [degree = 2]

```python
def poly_2(y):
    S = np.abs(librosa.stft(y[0]))
    p2 = librosa.feature.poly_features(S=S, order=2)
    return p2
```

```python
df['poly_2'] = df['y and SR'].apply(poly_2)
```

```python
df['poly_2 min'] = df['poly_2'].apply(lambda x: x.min())
df['poly_2 mean'] = df['poly_2'].apply(lambda x: x.mean())
df['poly_2 max'] = df['poly_2'].apply(lambda x: x.max())
```

```python
df
```

```python
def tonnetz(x):
    y = librosa.effects.harmonic(x)
    tonnetz = librosa.feature.tonnetz(y=y, sr=22050)
    return tonnetz
```

```python
df['tonnetz'] = df['y and SR'].apply(lambda x: librosa.feature.tonnetz(y=x[0], sr=x[1]))
```

```python
df['tonnetz min'] = df['tonnetz'].apply(lambda x: x.min())
df['tonnetz mean'] = df['tonnetz'].apply(lambda x: x.mean())
df['tonnetz max'] = df['tonnetz'].apply(lambda x: x.max())
```

```python
df['zero_crossing_rate'] = df['y'].apply(lambda x: librosa.feature.zero_crossing_rate(x))
```

```python
df['zero_crossing_rate min'] = df['zero_crossing_rate'].apply(lambda x: x.min())
df['zero_crossing_rate mean'] = df['zero_crossing_rate'].apply(lambda x: x.mean())
df['zero_crossing_rate max'] = df['zero_crossing_rate'].apply(lambda x: x.max())
```

```python
df.columns
```

Saving the dataframe with the features to a csv

```python
df1 = df[['Status', 'Type of Machine', 'File Name', 'File Path', 'Type of SNR', 'Model Number', 'SR', 'y mean', 'y max', 'y min', 
          'chroma_stft min', 'chroma_stft mean', 'chroma_stft max', 'chroma_cqt min', 'chroma_cqt mean', 'chroma_cqt max',
          'chroma_cens min', 'chroma_cens mean', 'chroma_cens max', 'melspectogram min', 'melspectogram mean',
          'melspectogram max',  'mfcc min', 'mfcc mean', 'mfcc max', 'rms min', 'rms mean', 'rms max', 'spectral_centroid min',
          'spectral_centroid mean', 'spectral_centroid max', 'spec_bw min', 'spec_bw mean', 'spec_bw max', 'spec_contrast min',
          'spec_contrast mean', 'spec_contrast max', 'flatness min', 'flatness mean', 'flatness max', 'rolloff min', 
          'rolloff mean', 'rolloff max', 'poly_0 min', 'poly_0 mean','poly_0 max',  'poly_1 min', 'poly_1 mean', 'poly_1 max',
          'poly_2 min', 'poly_2 mean', 'poly_2 max', 'tonnetz min', 'tonnetz mean', 'tonnetz max', 'zero_crossing_rate min',
          'zero_crossing_rate mean', 'zero_crossing_rate max']].copy()
```

```python
df1.to_csv("datasets/valve_6dB_features.csv")
```

```python
end_time = time.time()
print(f"The program runs for {end_time - start_time} seconds.")
```
